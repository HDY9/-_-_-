```python
import folium

folium.Map(location=[37.1325821,128.1909481], zoom_start=12)
```

```python
import pandas as pd
pd.read_excel('제천_여행지_목록.xlsx')
```

```python
df = pd.read_excel('제천_여행지_목록.xlsx')
```

```python
df.columns = ['구분','위도','경도']
```

```python
df.columns
```

```python
df.set_index('구분', inplace=True)
```

```python
df
```

```python
jaechun_map_terrain = folium.Map(location=[37.1325821,128.1909481], 
                               zoom_start=12)
```

```python
for name, lat, lon in zip(df.index, df.위도, df.경도): 
    print(name,lat,lon)
    folium.Marker([lat,lon], popup=name).add_to(jaechun_map_terrain)

folium.Marker(
    location = [37.128045, 128.2053],
    icon = folium.Icon(color='black',
                      icon='cloud',
                      icon_color='red'), tooltip='<i>제천_기차역</i>').add_to(jaechun_map_terrain)

folium.Marker(
    location = [37.1325069, 128.1204478],
    icon = folium.Icon(color='yellow',
                      icon='cloud',
                      icon_color='blue'), tooltip='<i>제천_팔송더덕막국수</i>').add_to(jaechun_map_terrain)
```
